```python
# Importación de librerías principales
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset

df_clientes = pd.read_csv('Raw/clientes.csv')

# Mostrar las primeras 5 filas para entender la estructura
display(df_clientes.head())
```

```python
# Información general: cantidad de filas, columnas, tipos de datos y valores no nulos
print("--- Información General del Dataset ---")
df_clientes.info()

print(f"\nDimensiones totales: {df_clientes.shape[0]} filas y {df_clientes.shape[1]} columnas.")
```

```python
# 1. Convertir la fecha a formato datetime
df_clientes['fecha_alta'] = pd.to_datetime(df_clientes['fecha_alta'])

df_clientes['fecha_alta'] = pd.to_datetime(
    df_clientes['fecha_alta'],
    format='mixed',
    dayfirst=True
)

# 2. Convertir el ID a string (object) porque es un identificador, no un número para operar
df_clientes['id_cliente'] = df_clientes['id_cliente'].astype(str)

# 3. Convertir a tipo 'category' las variables que tienen pocos valores únicos (ahorra memoria y agiliza cruces)
df_clientes['provincia'] = df_clientes['provincia'].astype('category')
df_clientes['segmento'] = df_clientes['segmento'].astype('category')

# Verificamos que los cambios se hayan aplicado
print("--- Nuevos Tipos de Datos ---")
display(df_clientes.dtypes)
```

```python
# Chequeo de valores nulos
print("--- Conteo de Valores Nulos ---")
display(df_clientes.isnull().sum())

# Chequeo de duplicados totales y por ID
duplicados_totales = df_clientes.duplicated().sum()
ids_duplicados = df_clientes.duplicated(subset=['id_cliente']).sum()

print("\n--- Análisis de Duplicados ---")
print(f"Filas exactamente duplicadas en todas las columnas: {duplicados_totales}")
print(f"IDs de cliente repetidos: {ids_duplicados}")

# Si hubiera IDs duplicados, podríamos verlos con:
# display(df_clientes[df_clientes.duplicated(subset=['id_cliente'], keep=False)].sort_values('id_cliente'))
```

```python
# Resumen descriptivo para variables categóricas y fechas
print("--- Resumen Estadístico (Categóricas y Fechas) ---")
display(df_clientes.describe(include=['category', 'object', 'datetime64']))
```
